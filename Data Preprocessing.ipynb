### Fill NULL values

**Fill values with Mean**

```python
#identify numeric columns

numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
numeric_columns
```

```python
#Filling values with Mean

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

**Fill values with Mode (Most Frequent)**

```python
#identify categorical columns

categorical_columns = ['Gender', 'Blood Type','Diabetes']
categorical_columns
```

```python
##Filling values with Mode

df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

**Drop non required rows**

```python
#Drop rows for NaN values in 'Smoking'

df.dropna(subset=['Smoking'], axis=0, inplace= True)
df
```

**Reset Index**

```python
df.reset_index(drop=True, inplace=True)
df
```

**Creating Dummies**

```python
convert_to_dummies = ['Gender','Blood Type', 'Diabetes']
```

```python
field_dummies = pd.get_dummies(df[convert_to_dummies])
field_dummies
```

**Add Dummies to code**

```python
df = pd.concat([df,field_dummies],axis = 1)
```

```python
df.drop(convert_to_dummies, axis=1, inplace=True)
df
```

# Building Model

**Split Data**

**Make X and Y**

```python
X = df.drop('Smoking',axis=1)
y = df['Smoking']

print("X: ", X.shape)
print("y: ", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
df
```

**Scale data**

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train[['Blood Pressure','Temperature']] = sc.fit_transform(X_train[['Blood Pressure','Temperature']])
X_test[['Blood Pressure','Temperature']] = sc.fit_transform(X_test[['Blood Pressure','Temperature']])
```

**Train the model**

```python
from sklearn.svm import SVC

model_linear = SVC(kernel='linear')
model_linear.fit(X_train[:500], y_train[:500])
```

**Get the accuracy**

```python
from sklearn import metrics
# predict
y_pred = model_linear.predict(X_test)

# accuracy
print("accuracy:", metrics.accuracy_score(y_test, y_pred), "\n")
```
